```python
import numpy as np
import matplotlib.pyplot as plt
R = 3
u = 10
def speed(x,y):
  return [3*(R**3)*u*(x**2)/(2*((x**2+y**2)**(5/2))) - R**3*u/(2*((x**2+y**2)**(3/2))), (3*u*(R**3)*x*y/((x**2+y**2)**(5/2)))]
dataframe = np.linspace(-10,10,20)
mas1 = [0,0]
fig = plt.subplots()
for x in (dataframe):
  for y in (dataframe):
    mas1 = speed(x,y)
    mas1[0] = mas1[0]/(((mas1[0])**2 + (mas1[1])**2)**(1/2))
    mas1[1] = mas1[1]/(((mas1[0])**2 + (mas1[1])**2)**(1/2))
    plt.arrow(x,y,mas1[0]/4, mas1[1]/4, width = 0.05)
dx = 1
dy = 1
x = np.arange(-10.0, 10.0, dx)
y = np.arange(-10.0, 10.0, dy)
extent = np.min(x), np.max(x), np.min(y)+1, np.max(y)+1
X, Y = np.meshgrid(x, y)
def geeks(x, y):
    return  ((3*(R**3)*u*(x**2)/(2*((x**2+y**2)**(5/2))) - R**3*u/(2*((x**2+y**2)**(3/2))))**2 + (3*u*(R**3)*x*y/((x**2+y**2)**(5/2)))**2)**(1/2)
Z2 = geeks(X, Y)
plt.imshow(Z2, alpha = 0.7, interpolation ='bilinear', extent = extent, vmin = 0, vmax = 20)
```

```python
p = 1
def pressure(x,y):
  return p*(R**3)*(u**2)*(3*(x**2) - (x**2+y**2))/(2*(x**2+y**2)**(5/2)) - p*(R**6)*(u**2)*(3*(x**2) + (x**2+y**2))/(8*(x**2+y**2)**(4)) + 2
dx = 1
dy = 1
x = np.arange(-10.0, 10.0, dx)
y = np.arange(-10.0, 10.0, dy)
extent = np.min(x), np.max(x), np.min(y)+1, np.max(y)+1
X, Y = np.meshgrid(x, y)
Z1 = pressure(X, Y)
plt.imshow(Z1, alpha = 0.7, interpolation ='bilinear', extent = extent, vmin = 0, vmax = 20)
```
